```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件

# 忽略警告訊息
%matplotlib inline
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')

# 設定 data_path
dir_data = './data/'
```

```python
# 讀取檔案
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
# 使用不同的 HOUSETYPE_MODE 類別繪製圖形, 並使用 subplot 排版
"""
Your Code Here
"""
data = app_train

unique_house_type = data.loc[list(~data['HOUSETYPE_MODE'].isnull()), 'HOUSETYPE_MODE'].unique()

num_rows = len(unique_house_type)
num_columns = num_rows // 2

column_name = 'AMT_INCOME_TOTAL'
excludePercentage = 0.05
clampMin = data[column_name].quantile(excludePercentage)
clampMax = data[column_name].quantile(1.0 - excludePercentage)
data[column_name] = data[column_name].clip(clampMin, clampMax)


plt.figure(figsize=(10,30))
for i in range(len(unique_house_type)):
    plt.subplot(num_rows, num_columns, i+1)
    data.loc[list(data['HOUSETYPE_MODE'] == unique_house_type[i]), column_name].hist()
    
    plt.title(str(unique_house_type[i]))

plt.show()

#nrows = len(unique_house_type)
#ncols = nrows // 2
#
#plt.figure(figsize=(10,30))
#for i in range(len(unique_house_type)):
#    plt.subplot(nrows, ncols, i+1)
#    """
#    Your Code Here
#    """
#    app_train.loc[ , ].hist()
#    
#    plt.title(str(unique_house_type[i]))
#plt.show()    
```
